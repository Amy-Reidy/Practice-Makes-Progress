```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
sns.set_palette('bright')
%matplotlib inline
```

```python
data = pd.read_csv('https://opendata.ecdc.europa.eu/covid19/casedistribution/csv')
```

```python
dataIRL = data.loc[data['countriesAndTerritories'] == 'Ireland']
dataIRL
```

```python
# plot cumulative number for 14 days of COVID 19 cases per 100000
plt.figure(figsize=(8,6))
plt.plot(dataIRL['month'], dataIRL['Cumulative_number_for_14_days_of_COVID-19_cases_per_100000'])
plt.xlabel('Month',fontsize=12, fontweight='bold')
plt.xticks(np.arange(min(dataIRL['month']), max(dataIRL['month'])+1, 1))
plt.ylabel('Number of Cases',fontsize=12, fontweight='bold')
plt.title('Cumulative Number for 14 days of COVID-19 cases per 100,000', fontsize=12, fontweight='bold')
```

```python

# plot cumulative number for 14 days of COVID 19 cases per 100000
plt.figure(figsize=(8,6))
sns.boxplot(dataIRL['month'], dataIRL['Cumulative_number_for_14_days_of_COVID-19_cases_per_100000'])
plt.xlabel('Month',fontsize=12, fontweight='bold')
plt.xticks(np.arange(min(dataIRL['month']), max(dataIRL['month'])+1, 1))
plt.ylabel('Number of Cases',fontsize=12, fontweight='bold')
plt.title('Cumulative Number for 14 days of COVID-19 cases per 100,000', fontsize=12, fontweight='bold')
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=dataIRL['month'],y=dataIRL.iloc[:,4], color='r')
plt.xlabel('Month',fontfamily='serif',fontsize=11, fontweight='bold')
plt.xticks(np.arange(min(dataIRL['month']), max(dataIRL['month'])+1, 1))
plt.ylabel('Number of Cases', fontfamily='serif',fontsize=11, fontweight='bold')
plt.title('Monthly Confirmed Cases in Ireland', fontfamily='serif', fontweight='bold')
```

```python
# COVID 19 Munster Counties Data
dataIRL = pd.read_csv("http://opendata-geohive.hub.arcgis.com/datasets/d9be85b30d7748b5b7c09450b8aede63_0.csv")
dataIRL
```

```python
counties = ['Kerry', 'Cork', 'Limerick', 'Waterford', 'Tipperary', 'Clare']
datamunster = dataIRL[dataIRL['CountyName'].isin(counties)]
```

```python
from datetime import datetime

# split time and date
datamunster['Timestamp'] = pd.to_datetime(datamunster['TimeStamp']).dt.date

# convert date to Month Day format
datamunster['TimeStamp'] = pd.to_datetime(datamunster['TimeStamp']).dt.strftime('%d-%b-%y')

datamunster
```

```python
import matplotlib.dates as mdates

plt.figure(figsize=(8,6))
ax = plt.gca()

ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))

groups = datamunster.groupby('CountyName')

for name, group in groups:
    ax.scatter(group['TimeStamp'], group['ConfirmedCovidCases'], label=name, linestyle='dotted', s=10)
    
plt.setp(ax.get_xticklabels(), rotation=45, fontsize=12)
ax.set_xlabel('Date')
ax.set_ylabel('Number of Confirmed Cases')
ax.set_title('Number of Confirmed Cases in Munster Counties', fontweight='bold', fontsize=20)
ax.legend(title='County', ncol=2)
```
